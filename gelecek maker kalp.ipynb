```python
dizi_g = [0, 0, 1, 1, 1]
dizi_t = [0, 0, 1, 1, 0]

TP = 0
TN = 0
FP = 0
FN = 0

for i in range(len(dizi_g)):
    if dizi_g[i-1] == dizi_t[i-1] and dizi_g[i-1] == 1:
        TP += 1
    if dizi_g[i-1] == dizi_t[i-1] and dizi_g[i-1] == 0:
        TN += 1
    if dizi_g[i-1] != dizi_t[i-1] and dizi_g[i-1] == 0:
        FN += 1
    if dizi_g[i-1] != dizi_t[i-1] and dizi_g[i-1] == 1:
        FP += 1

pre = TP / (TP + FP)
rec = TP / (TP + FN)

print("Accuracy:", (TP + TN) / (TP + TN + FP + FN))
print("Precision:", pre)
print("Recall:", rec )
print("F1 score:", 2 * (pre * rec) / (pre + rec))
```
